# Ablation Study

## Group block removal

```python
# Cell 1: Imports and Device Configuration
import os
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import glob
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score
import pandas as pd
import matplotlib.pyplot as plt

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Cell 2: Ablation Study Models (Removing Blocks)
class CompactOCTCNN_Full(nn.Module):
    """Full model with all 4 blocks (Original)"""
    def __init__(self, input_channels=1, num_classes=2):
        super(CompactOCTCNN_Full, self).__init__()
        
        self.conv_layers = nn.Sequential(
            nn.Conv2d(input_channels, 8, kernel_size=3, padding=1),
            nn.BatchNorm2d(8),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            nn.Conv2d(8, 16, kernel_size=3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 4))
        )
        
        self.fc_layers = nn.Sequential(
            nn.Linear(64 * 4 * 4, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, num_classes)
        )
        
    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.size(0), -1)
        x = self.fc_layers(x)
        return x

class CompactOCTCNN_NoBlock4(nn.Module):
    """Remove Block 4 (no 64-channel conv layer)"""
    def __init__(self, input_channels=1, num_classes=2):
        super(CompactOCTCNN_NoBlock4, self).__init__()
        
        self.conv_layers = nn.Sequential(
            nn.Conv2d(input_channels, 8, kernel_size=3, padding=1),
            nn.BatchNorm2d(8),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            nn.Conv2d(8, 16, kernel_size=3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 4))  # Removed Block 4
        )
        
        # Adjusted FC layer (32 channels instead of 64)
        self.fc_layers = nn.Sequential(
            nn.Linear(32 * 4 * 4, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, num_classes)
        )
        
    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.size(0), -1)
        x = self.fc_layers(x)
        return x

class CompactOCTCNN_NoBlock3_4(nn.Module):
    """Remove Blocks 3 and 4 (stop at 16 channels)"""
    def __init__(self, input_channels=1, num_classes=2):
        super(CompactOCTCNN_NoBlock3_4, self).__init__()
        
        self.conv_layers = nn.Sequential(
            nn.Conv2d(input_channels, 8, kernel_size=3, padding=1),
            nn.BatchNorm2d(8),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            nn.Conv2d(8, 16, kernel_size=3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 4))  # Removed Blocks 3 and 4
        )
        
        # Adjusted FC layer (16 channels)
        self.fc_layers = nn.Sequential(
            nn.Linear(16 * 4 * 4, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, num_classes)
        )
        
    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.size(0), -1)
        x = self.fc_layers(x)
        return x

class CompactOCTCNN_NoBlock2_3_4(nn.Module):
    """Remove Blocks 2, 3, and 4 (only Block 1 - 8 channels)"""
    def __init__(self, input_channels=1, num_classes=2):
        super(CompactOCTCNN_NoBlock2_3_4, self).__init__()
        
        self.conv_layers = nn.Sequential(
            nn.Conv2d(input_channels, 8, kernel_size=3, padding=1),
            nn.BatchNorm2d(8),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 4))  # Only Block 1
        )
        
        # Adjusted FC layer (8 channels)
        self.fc_layers = nn.Sequential(
            nn.Linear(8 * 4 * 4, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, num_classes)
        )
        
    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.size(0), -1)
        x = self.fc_layers(x)
        return x

# Cell 3: Dataset and Data Loading Functions (Same as before)
class FrameLevelDataset(Dataset):
    def __init__(self, images, labels, patient_ids=None):
        self.images = images
        self.labels = labels
        self.patient_ids = patient_ids
        
    def __len__(self):
        return len(self.images)
    
    def __getitem__(self, idx):
        image = self.images[idx]
        label = self.labels[idx]
        
        if len(image.shape) == 2:
            image = np.expand_dims(image, axis=0)
        
        image = torch.FloatTensor(image)
        return image, label

def load_and_preprocess_image(img_path, img_size=64):
    img = Image.open(img_path)
    if img.mode != 'L':
        img = img.convert('L')
    img = img.resize((img_size, img_size))
    img_array = np.array(img, dtype=np.float32) / 255.0
    return img_array

def load_frame_data_with_patient_info(data_dir, img_size=64):
    images = []
    labels = []
    patient_ids = []
    image_paths = []
    
    class_mapping = {'MS': 1, 'Control': 0}
    
    for class_name, class_label in class_mapping.items():
        class_dir = os.path.join(data_dir, class_name)
        if not os.path.exists(class_dir):
            print(f"Warning: {class_dir} does not exist")
            continue
            
        image_files = glob.glob(os.path.join(class_dir, "*.png")) + \
                     glob.glob(os.path.join(class_dir, "*.jpg")) + \
                     glob.glob(os.path.join(class_dir, "*.jpeg"))
        image_files.sort()
        
        print(f"Loading {class_name} data: {len(image_files)} frames")
        
        for img_path in image_files:
            try:
                img = load_and_preprocess_image(img_path, img_size)
                images.append(img)
                labels.append(class_label)
                image_paths.append(img_path)
                
                filename = os.path.basename(img_path)
                patient_id = filename.split('_')[0]
                patient_ids.append(patient_id)
                
            except Exception as e:
                print(f"Error loading {img_path}: {e}")
                continue
    
    return np.array(images), np.array(labels), patient_ids, image_paths

# Cell 4: Training and Evaluation Classes
class OCTTrainer:
    def __init__(self, model, device, criterion, optimizer):
        self.model = model
        self.device = device
        self.criterion = criterion
        self.optimizer = optimizer
        self.train_losses = []
        self.val_losses = []
        self.train_accuracies = []
        self.val_accuracies = []
        
    def train_epoch(self, dataloader):
        self.model.train()
        running_loss = 0.0
        correct = 0
        total = 0
        
        for data, target in dataloader:
            data, target = data.to(self.device), target.to(self.device)
            
            self.optimizer.zero_grad()
            output = self.model(data)
            loss = self.criterion(output, target)
            loss.backward()
            self.optimizer.step()
            
            running_loss += loss.item()
            _, predicted = torch.max(output.data, 1)
            total += target.size(0)
            correct += (predicted == target).sum().item()
        
        epoch_loss = running_loss / len(dataloader)
        epoch_acc = 100. * correct / total
        return epoch_loss, epoch_acc
    
    def validate_epoch(self, dataloader):
        self.model.eval()
        running_loss = 0.0
        correct = 0
        total = 0
        
        with torch.no_grad():
            for data, target in dataloader:
                data, target = data.to(self.device), target.to(self.device)
                output = self.model(data)
                loss = self.criterion(output, target)
                
                running_loss += loss.item()
                _, predicted = torch.max(output.data, 1)
                total += target.size(0)
                correct += (predicted == target).sum().item()
        
        epoch_loss = running_loss / len(dataloader)
        epoch_acc = 100. * correct / total
        return epoch_loss, epoch_acc
    
    def train(self, train_loader, val_loader, epochs, verbose=False):
        for epoch in range(epochs):
            train_loss, train_acc = self.train_epoch(train_loader)
            val_loss, val_acc = self.validate_epoch(val_loader)
            
            self.train_losses.append(train_loss)
            self.val_losses.append(val_loss)
            self.train_accuracies.append(train_acc)
            self.val_accuracies.append(val_acc)
            
            if verbose:
                print(f'Epoch {epoch+1}/{epochs}:')
                print(f'  Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%')
                print(f'  Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%')

class ModelEvaluator:
    @staticmethod
    def evaluate_model(model, dataloader, device):
        model.eval()
        all_predictions = []
        all_targets = []
        all_probabilities = []
        
        with torch.no_grad():
            for data, target in dataloader:
                data, target = data.to(device), target.to(device)
                output = model(data)
                probabilities = torch.softmax(output, dim=1)
                _, predicted = torch.max(output, 1)
                
                all_predictions.extend(predicted.cpu().numpy())
                all_targets.extend(target.cpu().numpy())
                all_probabilities.extend(probabilities.cpu().numpy())
        
        return np.array(all_predictions), np.array(all_targets), np.array(all_probabilities)
    
    @staticmethod
    def calculate_metrics(predictions, targets, probabilities):
        accuracy = accuracy_score(targets, predictions)
        precision = precision_score(targets, predictions, average='binary', zero_division=0)
        recall = recall_score(targets, predictions, average='binary', zero_division=0)
        f1 = f1_score(targets, predictions, average='binary', zero_division=0)
        
        cm = confusion_matrix(targets, predictions)
        
        if cm.shape == (2, 2):
            tn, fp, fn, tp = cm.ravel()
            sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
            specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
        else:
            sensitivity = recall
            specificity = 0
        
        auc = roc_auc_score(targets, probabilities[:, 1]) if len(np.unique(targets)) > 1 else 0.5
        
        metrics = {
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1_score': f1,
            'sensitivity': sensitivity,
            'specificity': specificity,
            'auc': auc,
            'confusion_matrix': cm
        }
        
        return metrics

# Cell 5: Ablation Study LOOCV Function
def run_ablation_study(data_dir, img_size=64, epochs=15, batch_size=8):
    """
    Run ablation study by removing blocks one by one
    """
    print("="*80)
    print("ABLATION STUDY: Removing Blocks from 2D CNN")
    print("="*80)
    
    # Load data once
    images, labels, patient_ids, image_paths = load_frame_data_with_patient_info(data_dir, img_size)
    
    if len(images) == 0:
        print("No data loaded!")
        return None
    
    # Get unique patients
    unique_patients = list(set(patient_ids))
    print(f"\nFound {len(unique_patients)} unique patients")
    print(f"Total frames: {len(images)}")
    
    # Define models for ablation study
    models = {
        'Full Model (4 Blocks)': CompactOCTCNN_Full,
        'Remove Block 4 (64→32 channels)': CompactOCTCNN_NoBlock4,
        'Remove Blocks 3-4 (32→16 channels)': CompactOCTCNN_NoBlock3_4,
        'Remove Blocks 2-3-4 (16→8 channels)': CompactOCTCNN_NoBlock2_3_4
    }
    
    # Store results
    ablation_results = {}
    
    for model_name, model_class in models.items():
        print(f"\n{'='*70}")
        print(f"TESTING: {model_name}")
        print(f"{'='*70}")
        
        # Initialize model
        model = model_class(input_channels=1, num_classes=2)
        total_params = sum(p.numel() for p in model.parameters())
        print(f"Total parameters: {total_params:,}")
        
        # Store results for this model
        all_predictions = []
        all_targets = []
        all_probabilities = []
        patient_level_results = {}
        
        # Leave-One-Patient-Out Cross Validation
        for fold, test_patient in enumerate(unique_patients):
            # Split data by patient
            train_indices = [i for i, pid in enumerate(patient_ids) if pid != test_patient]
            test_indices = [i for i, pid in enumerate(patient_ids) if pid == test_patient]
            
            if len(train_indices) == 0 or len(test_indices) == 0:
                continue
            
            # Get train and test data
            X_train = images[train_indices]
            y_train = labels[train_indices]
            X_test = images[test_indices]
            y_test = labels[test_indices]
            
            # Split training data for validation
            train_patients = list(set([patient_ids[i] for i in train_indices]))
            if len(train_patients) > 1:
                val_patient = train_patients[0]
                val_indices = [i for i, pid in enumerate(patient_ids) if pid == val_patient and i in train_indices]
                train_indices_final = [i for i in train_indices if patient_ids[i] != val_patient]
                
                X_train_final = images[train_indices_final]
                y_train_final = labels[train_indices_final]
                X_val = images[val_indices]
                y_val = labels[val_indices]
            else:
                X_train_final, X_val, y_train_final, y_val = X_train, X_train, y_train, y_train
            
            # Create data loaders
            train_dataset = FrameLevelDataset(X_train_final, y_train_final)
            val_dataset = FrameLevelDataset(X_val, y_val)
            test_dataset = FrameLevelDataset(X_test, y_test)
            
            train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
            val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
            test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
            
            # Reinitialize model for each fold
            model = model_class(input_channels=1, num_classes=2)
            model.to(device)
            
            criterion = nn.CrossEntropyLoss()
            optimizer = optim.Adam(model.parameters(), lr=0.0001, weight_decay=1e-4)
            
            # Train model
            trainer = OCTTrainer(model, device, criterion, optimizer)
            trainer.train(train_loader, val_loader, epochs, verbose=False)
            
            # Evaluate on test set
            predictions, targets, probabilities = ModelEvaluator.evaluate_model(model, test_loader, device)
            
            if len(targets) > 0:
                all_predictions.extend(predictions)
                all_targets.extend(targets)
                all_probabilities.extend(probabilities)
                
                # Store patient-level metrics
                patient_metrics = ModelEvaluator.calculate_metrics(predictions, targets, probabilities)
                patient_level_results[test_patient] = patient_metrics
        
        # Calculate overall metrics for this model
        if len(all_targets) > 0:
            overall_metrics = ModelEvaluator.calculate_metrics(
                np.array(all_predictions), 
                np.array(all_targets), 
                np.array(all_probabilities)
            )
            
            ablation_results[model_name] = {
                'overall_metrics': overall_metrics,
                'patient_results': patient_level_results,
                'total_params': total_params
            }
            
            # Print results
            print(f"\nResults for {model_name}:")
            print(f"  Accuracy:    {overall_metrics['accuracy']:.4f}")
            print(f"  Precision:   {overall_metrics['precision']:.4f}")
            print(f"  Recall:      {overall_metrics['recall']:.4f}")
            print(f"  F1-Score:    {overall_metrics['f1_score']:.4f}")
            print(f"  Sensitivity: {overall_metrics['sensitivity']:.4f}")
            print(f"  Specificity: {overall_metrics['specificity']:.4f}")
            print(f"  AUC:         {overall_metrics['auc']:.4f}")
    
    return ablation_results

# Cell 6: Visualization and Comparison
def visualize_ablation_results(ablation_results):
    """
    Create visualizations for ablation study results
    """
    if not ablation_results:
        print("No results to visualize!")
        return
    
    # Extract metrics for comparison
    model_names = list(ablation_results.keys())
    metrics_names = ['accuracy', 'precision', 'recall', 'f1_score', 'sensitivity', 'specificity', 'auc']
    
    # Create DataFrame
    results_df = pd.DataFrame()
    for model_name in model_names:
        metrics = ablation_results[model_name]['overall_metrics']
        params = ablation_results[model_name]['total_params']
        row = {'Model': model_name, 'Parameters': params}
        for metric in metrics_names:
            row[metric.capitalize()] = metrics[metric]
        results_df = pd.concat([results_df, pd.DataFrame([row])], ignore_index=True)
    
    # Print comparison table
    print("\n" + "="*100)
    print("ABLATION STUDY RESULTS SUMMARY")
    print("="*100)
    print(results_df.to_string(index=False))
    
    # Create bar plots
    fig, axes = plt.subplots(2, 4, figsize=(20, 10))
    axes = axes.flatten()
    
    for idx, metric in enumerate(metrics_names):
        ax = axes[idx]
        metric_values = [ablation_results[model]['overall_metrics'][metric] for model in model_names]
        
        bars = ax.bar(range(len(model_names)), metric_values, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
        ax.set_title(f'{metric.capitalize()} Comparison', fontsize=14, fontweight='bold')
        ax.set_xlabel('Model Configuration', fontsize=12)
        ax.set_ylabel(metric.capitalize(), fontsize=12)
        ax.set_xticks(range(len(model_names)))
        ax.set_xticklabels(['Full', 'No Block4', 'No B3-4', 'No B2-3-4'], rotation=15, ha='right')
        ax.set_ylim([0, 1.1])
        
        # Add value labels on bars
        for bar, value in zip(bars, metric_values):
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., height + 0.02,
                   f'{value:.3f}', ha='center', va='bottom', fontsize=10, fontweight='bold')
        
        ax.grid(True, alpha=0.3, axis='y')
    
    # Plot parameter comparison
    ax = axes[7]
    param_values = [ablation_results[model]['total_params'] for model in model_names]
    bars = ax.bar(range(len(model_names)), param_values, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
    ax.set_title('Model Parameters Comparison', fontsize=14, fontweight='bold')
    ax.set_xlabel('Model Configuration', fontsize=12)
    ax.set_ylabel('Number of Parameters', fontsize=12)
    ax.set_xticks(range(len(model_names)))
    ax.set_xticklabels(['Full', 'No Block4', 'No B3-4', 'No B2-3-4'], rotation=15, ha='right')
    
    # Add value labels
    for bar, value in zip(bars, param_values):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + height*0.02,
               f'{value:,}', ha='center', va='bottom', fontsize=10, fontweight='bold')
    
    ax.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.savefig('ablation_study_results.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Calculate performance drop
    print("\n" + "="*100)
    print("PERFORMANCE DROP ANALYSIS")
    print("="*100)
    
    full_metrics = ablation_results[model_names[0]]['overall_metrics']
    
    for model_name in model_names[1:]:
        print(f"\n{model_name}:")
        current_metrics = ablation_results[model_name]['overall_metrics']
        for metric in metrics_names:
            drop = (full_metrics[metric] - current_metrics[metric]) * 100
            print(f"  {metric.capitalize()}: {current_metrics[metric]:.4f} (Drop: {drop:.2f}%)")
    
    return results_df

# Cell 7: Main Function for Ablation Study
def main():
    """Main function to run ablation study"""
    print("OCT Classification - Ablation Study on 2D CNN")
    print("=" * 70)
    
    # UPDATE THIS PATH to your frame-level data directory
    FRAME_LEVEL_DATA_PATH = "/home/tanvirdell3/Downloads/caserel-master/Publically available Dataset/Segmentation Sharpen filter"
    
    if not os.path.exists(FRAME_LEVEL_DATA_PATH):
        print(f"Data path {FRAME_LEVEL_DATA_PATH} does not exist!")
        return
    
    # Run ablation study
    ablation_results = run_ablation_study(
        FRAME_LEVEL_DATA_PATH,
        img_size=64,
        epochs=50,
        batch_size=8
    )
    
    # Visualize results
    if ablation_results:
        results_df = visualize_ablation_results(ablation_results)
        
        # Save results to CSV
        results_df.to_csv('ablation_study_results.csv', index=False)
        print("\nResults saved to 'ablation_study_results.csv'")
        
        # Save detailed results
        with open('ablation_study_detailed.txt', 'w') as f:
            f.write("ABLATION STUDY DETAILED RESULTS\n")
            f.write("="*80 + "\n\n")
            for model_name, results in ablation_results.items():
                f.write(f"{model_name}\n")
                f.write(f"Parameters: {results['total_params']:,}\n")
                f.write("-"*40 + "\n")
                for metric, value in results['overall_metrics'].items():
                    if metric != 'confusion_matrix':
                        f.write(f"{metric}: {value:.4f}\n")
                f.write(f"\nConfusion Matrix:\n{results['overall_metrics']['confusion_matrix']}\n")
                f.write("\n" + "="*80 + "\n\n")
        
        print("Detailed results saved to 'ablation_study_detailed.txt'")

# Run the main function
if __name__ == "__main__":
    main()
```

## One By one Removal

```python
# Cell 1: Imports and Device Configuration
import os
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import glob
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score
import pandas as pd
import matplotlib.pyplot as plt

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Cell 2: Ablation Study Models (Remove ONE Block at a time)

class CompactOCTCNN_Full(nn.Module):
    """Full model with all 4 blocks (Original)"""
    def __init__(self, input_channels=1, num_classes=2):
        super(CompactOCTCNN_Full, self).__init__()
        
        self.conv_layers = nn.Sequential(
            # Block 1
            nn.Conv2d(input_channels, 8, kernel_size=3, padding=1),
            nn.BatchNorm2d(8),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            # Block 2
            nn.Conv2d(8, 16, kernel_size=3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            # Block 3
            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            # Block 4
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 4))
        )
        
        self.fc_layers = nn.Sequential(
            nn.Linear(64 * 4 * 4, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, num_classes)
        )
        
    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.size(0), -1)
        x = self.fc_layers(x)
        return x

class CompactOCTCNN_RemoveBlock2(nn.Module):
    """Remove Block 2 only (Keep Blocks 1, 3, 4)"""
    def __init__(self, input_channels=1, num_classes=2):
        super(CompactOCTCNN_RemoveBlock2, self).__init__()
        
        self.conv_layers = nn.Sequential(
            # Block 1
            nn.Conv2d(input_channels, 8, kernel_size=3, padding=1),
            nn.BatchNorm2d(8),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            # Block 2 - REMOVED (skip 8→16 directly to 32)
            
            # Block 3 (now takes 8 channels directly)
            nn.Conv2d(8, 32, kernel_size=3, padding=1),  # Note: 8→32 directly
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            # Block 4
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 4))
        )
        
        self.fc_layers = nn.Sequential(
            nn.Linear(64 * 4 * 4, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, num_classes)
        )
        
    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.size(0), -1)
        x = self.fc_layers(x)
        return x

class CompactOCTCNN_RemoveBlock3(nn.Module):
    """Remove Block 3 only (Keep Blocks 1, 2, 4)"""
    def __init__(self, input_channels=1, num_classes=2):
        super(CompactOCTCNN_RemoveBlock3, self).__init__()
        
        self.conv_layers = nn.Sequential(
            # Block 1
            nn.Conv2d(input_channels, 8, kernel_size=3, padding=1),
            nn.BatchNorm2d(8),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            # Block 2
            nn.Conv2d(8, 16, kernel_size=3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            # Block 3 - REMOVED (skip 16→32 directly to 64)
            
            # Block 4 (now takes 16 channels directly)
            nn.Conv2d(16, 64, kernel_size=3, padding=1),  # Note: 16→64 directly
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 4))
        )
        
        self.fc_layers = nn.Sequential(
            nn.Linear(64 * 4 * 4, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, num_classes)
        )
        
    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.size(0), -1)
        x = self.fc_layers(x)
        return x

class CompactOCTCNN_RemoveBlock4(nn.Module):
    """Remove Block 4 only (Keep Blocks 1, 2, 3)"""
    def __init__(self, input_channels=1, num_classes=2):
        super(CompactOCTCNN_RemoveBlock4, self).__init__()
        
        self.conv_layers = nn.Sequential(
            # Block 1
            nn.Conv2d(input_channels, 8, kernel_size=3, padding=1),
            nn.BatchNorm2d(8),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            # Block 2
            nn.Conv2d(8, 16, kernel_size=3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            # Block 3
            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 4))  # Removed Block 4, so adaptive pool here
        )
        
        # FC layer with 32 channels instead of 64
        self.fc_layers = nn.Sequential(
            nn.Linear(32 * 4 * 4, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, num_classes)
        )
        
    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.size(0), -1)
        x = self.fc_layers(x)
        return x

# Cell 3: Dataset and Data Loading Functions
class FrameLevelDataset(Dataset):
    def __init__(self, images, labels, patient_ids=None):
        self.images = images
        self.labels = labels
        self.patient_ids = patient_ids
        
    def __len__(self):
        return len(self.images)
    
    def __getitem__(self, idx):
        image = self.images[idx]
        label = self.labels[idx]
        
        if len(image.shape) == 2:
            image = np.expand_dims(image, axis=0)
        
        image = torch.FloatTensor(image)
        return image, label

def load_and_preprocess_image(img_path, img_size=64):
    img = Image.open(img_path)
    if img.mode != 'L':
        img = img.convert('L')
    img = img.resize((img_size, img_size))
    img_array = np.array(img, dtype=np.float32) / 255.0
    return img_array

def load_frame_data_with_patient_info(data_dir, img_size=64):
    images = []
    labels = []
    patient_ids = []
    image_paths = []
    
    class_mapping = {'MS': 1, 'Control': 0}
    
    for class_name, class_label in class_mapping.items():
        class_dir = os.path.join(data_dir, class_name)
        if not os.path.exists(class_dir):
            print(f"Warning: {class_dir} does not exist")
            continue
            
        image_files = glob.glob(os.path.join(class_dir, "*.png")) + \
                     glob.glob(os.path.join(class_dir, "*.jpg")) + \
                     glob.glob(os.path.join(class_dir, "*.jpeg"))
        image_files.sort()
        
        print(f"Loading {class_name} data: {len(image_files)} frames")
        
        for img_path in image_files:
            try:
                img = load_and_preprocess_image(img_path, img_size)
                images.append(img)
                labels.append(class_label)
                image_paths.append(img_path)
                
                filename = os.path.basename(img_path)
                patient_id = filename.split('_')[0]
                patient_ids.append(patient_id)
                
            except Exception as e:
                print(f"Error loading {img_path}: {e}")
                continue
    
    return np.array(images), np.array(labels), patient_ids, image_paths

# Cell 4: Training and Evaluation Classes
class OCTTrainer:
    def __init__(self, model, device, criterion, optimizer):
        self.model = model
        self.device = device
        self.criterion = criterion
        self.optimizer = optimizer
        self.train_losses = []
        self.val_losses = []
        self.train_accuracies = []
        self.val_accuracies = []
        
    def train_epoch(self, dataloader):
        self.model.train()
        running_loss = 0.0
        correct = 0
        total = 0
        
        for data, target in dataloader:
            data, target = data.to(self.device), target.to(self.device)
            
            self.optimizer.zero_grad()
            output = self.model(data)
            loss = self.criterion(output, target)
            loss.backward()
            self.optimizer.step()
            
            running_loss += loss.item()
            _, predicted = torch.max(output.data, 1)
            total += target.size(0)
            correct += (predicted == target).sum().item()
        
        epoch_loss = running_loss / len(dataloader)
        epoch_acc = 100. * correct / total
        return epoch_loss, epoch_acc
    
    def validate_epoch(self, dataloader):
        self.model.eval()
        running_loss = 0.0
        correct = 0
        total = 0
        
        with torch.no_grad():
            for data, target in dataloader:
                data, target = data.to(self.device), target.to(self.device)
                output = self.model(data)
                loss = self.criterion(output, target)
                
                running_loss += loss.item()
                _, predicted = torch.max(output.data, 1)
                total += target.size(0)
                correct += (predicted == target).sum().item()
        
        epoch_loss = running_loss / len(dataloader)
        epoch_acc = 100. * correct / total
        return epoch_loss, epoch_acc
    
    def train(self, train_loader, val_loader, epochs, verbose=False):
        for epoch in range(epochs):
            train_loss, train_acc = self.train_epoch(train_loader)
            val_loss, val_acc = self.validate_epoch(val_loader)
            
            self.train_losses.append(train_loss)
            self.val_losses.append(val_loss)
            self.train_accuracies.append(train_acc)
            self.val_accuracies.append(val_acc)
            
            if verbose:
                print(f'Epoch {epoch+1}/{epochs}:')
                print(f'  Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%')
                print(f'  Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%')

class ModelEvaluator:
    @staticmethod
    def evaluate_model(model, dataloader, device):
        model.eval()
        all_predictions = []
        all_targets = []
        all_probabilities = []
        
        with torch.no_grad():
            for data, target in dataloader:
                data, target = data.to(device), target.to(device)
                output = model(data)
                probabilities = torch.softmax(output, dim=1)
                _, predicted = torch.max(output, 1)
                
                all_predictions.extend(predicted.cpu().numpy())
                all_targets.extend(target.cpu().numpy())
                all_probabilities.extend(probabilities.cpu().numpy())
        
        return np.array(all_predictions), np.array(all_targets), np.array(all_probabilities)
    
    @staticmethod
    def calculate_metrics(predictions, targets, probabilities):
        accuracy = accuracy_score(targets, predictions)
        precision = precision_score(targets, predictions, average='binary', zero_division=0)
        recall = recall_score(targets, predictions, average='binary', zero_division=0)
        f1 = f1_score(targets, predictions, average='binary', zero_division=0)
        
        cm = confusion_matrix(targets, predictions)
        
        if cm.shape == (2, 2):
            tn, fp, fn, tp = cm.ravel()
            sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
            specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
        else:
            sensitivity = recall
            specificity = 0
        
        auc = roc_auc_score(targets, probabilities[:, 1]) if len(np.unique(targets)) > 1 else 0.5
        
        metrics = {
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1_score': f1,
            'sensitivity': sensitivity,
            'specificity': specificity,
            'auc': auc,
            'confusion_matrix': cm
        }
        
        return metrics

# Cell 5: Ablation Study LOOCV Function (Remove ONE Block at a time)
def run_ablation_study(data_dir, img_size=64, epochs=15, batch_size=8):
    """
    Run ablation study by removing ONE block at a time
    """
    print("="*80)
    print("ABLATION STUDY: Removing ONE Block at a Time from 2D CNN")
    print("="*80)
    
    # Load data once
    images, labels, patient_ids, image_paths = load_frame_data_with_patient_info(data_dir, img_size)
    
    if len(images) == 0:
        print("No data loaded!")
        return None
    
    # Get unique patients
    unique_patients = list(set(patient_ids))
    print(f"\nFound {len(unique_patients)} unique patients")
    print(f"Total frames: {len(images)}")
    
    # Define models for ablation study (ONE block removed at a time)
    models = {
        'Full Model (All Blocks)': CompactOCTCNN_Full,
        'Remove Block 2 (Keep 1,3,4)': CompactOCTCNN_RemoveBlock2,
        'Remove Block 3 (Keep 1,2,4)': CompactOCTCNN_RemoveBlock3,
        'Remove Block 4 (Keep 1,2,3)': CompactOCTCNN_RemoveBlock4
    }
    
    # Store results
    ablation_results = {}
    
    for model_name, model_class in models.items():
        print(f"\n{'='*70}")
        print(f"TESTING: {model_name}")
        print(f"{'='*70}")
        
        # Initialize model
        model = model_class(input_channels=1, num_classes=2)
        total_params = sum(p.numel() for p in model.parameters())
        print(f"Total parameters: {total_params:,}")
        
        # Store results for this model
        all_predictions = []
        all_targets = []
        all_probabilities = []
        patient_level_results = {}
        
        # Leave-One-Patient-Out Cross Validation
        for fold, test_patient in enumerate(unique_patients):
            # Split data by patient
            train_indices = [i for i, pid in enumerate(patient_ids) if pid != test_patient]
            test_indices = [i for i, pid in enumerate(patient_ids) if pid == test_patient]
            
            if len(train_indices) == 0 or len(test_indices) == 0:
                continue
            
            # Get train and test data
            X_train = images[train_indices]
            y_train = labels[train_indices]
            X_test = images[test_indices]
            y_test = labels[test_indices]
            
            # Split training data for validation
            train_patients = list(set([patient_ids[i] for i in train_indices]))
            if len(train_patients) > 1:
                val_patient = train_patients[0]
                val_indices = [i for i, pid in enumerate(patient_ids) if pid == val_patient and i in train_indices]
                train_indices_final = [i for i in train_indices if patient_ids[i] != val_patient]
                
                X_train_final = images[train_indices_final]
                y_train_final = labels[train_indices_final]
                X_val = images[val_indices]
                y_val = labels[val_indices]
            else:
                X_train_final, X_val, y_train_final, y_val = X_train, X_train, y_train, y_train
            
            # Create data loaders
            train_dataset = FrameLevelDataset(X_train_final, y_train_final)
            val_dataset = FrameLevelDataset(X_val, y_val)
            test_dataset = FrameLevelDataset(X_test, y_test)
            
            train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
            val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
            test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
            
            # Reinitialize model for each fold
            model = model_class(input_channels=1, num_classes=2)
            model.to(device)
            
            criterion = nn.CrossEntropyLoss()
            optimizer = optim.Adam(model.parameters(), lr=0.0001, weight_decay=1e-4)
            
            # Train model
            trainer = OCTTrainer(model, device, criterion, optimizer)
            trainer.train(train_loader, val_loader, epochs, verbose=False)
            
            # Evaluate on test set
            predictions, targets, probabilities = ModelEvaluator.evaluate_model(model, test_loader, device)
            
            if len(targets) > 0:
                all_predictions.extend(predictions)
                all_targets.extend(targets)
                all_probabilities.extend(probabilities)
                
                # Store patient-level metrics
                patient_metrics = ModelEvaluator.calculate_metrics(predictions, targets, probabilities)
                patient_level_results[test_patient] = patient_metrics
        
        # Calculate overall metrics for this model
        if len(all_targets) > 0:
            overall_metrics = ModelEvaluator.calculate_metrics(
                np.array(all_predictions), 
                np.array(all_targets), 
                np.array(all_probabilities)
            )
            
            ablation_results[model_name] = {
                'overall_metrics': overall_metrics,
                'patient_results': patient_level_results,
                'total_params': total_params
            }
            
            # Print results
            print(f"\nResults for {model_name}:")
            print(f"  Accuracy:    {overall_metrics['accuracy']:.4f}")
            print(f"  Precision:   {overall_metrics['precision']:.4f}")
            print(f"  Recall:      {overall_metrics['recall']:.4f}")
            print(f"  F1-Score:    {overall_metrics['f1_score']:.4f}")
            print(f"  Sensitivity: {overall_metrics['sensitivity']:.4f}")
            print(f"  Specificity: {overall_metrics['specificity']:.4f}")
            print(f"  AUC:         {overall_metrics['auc']:.4f}")
    
    return ablation_results

# Cell 6: Visualization and Comparison
def visualize_ablation_results(ablation_results):
    """
    Create visualizations for ablation study results
    """
    if not ablation_results:
        print("No results to visualize!")
        return
    
    # Extract metrics for comparison
    model_names = list(ablation_results.keys())
    metrics_names = ['accuracy', 'precision', 'recall', 'f1_score', 'sensitivity', 'specificity', 'auc']
    
    # Create DataFrame
    results_df = pd.DataFrame()
    for model_name in model_names:
        metrics = ablation_results[model_name]['overall_metrics']
        params = ablation_results[model_name]['total_params']
        row = {'Model': model_name, 'Parameters': params}
        for metric in metrics_names:
            row[metric.capitalize()] = metrics[metric]
        results_df = pd.concat([results_df, pd.DataFrame([row])], ignore_index=True)
    
    # Print comparison table
    print("\n" + "="*100)
    print("ABLATION STUDY RESULTS SUMMARY (Remove ONE Block at a Time)")
    print("="*100)
    print(results_df.to_string(index=False))
    
    # Create bar plots
    fig, axes = plt.subplots(2, 4, figsize=(20, 10))
    axes = axes.flatten()
    
    for idx, metric in enumerate(metrics_names):
        ax = axes[idx]
        metric_values = [ablation_results[model]['overall_metrics'][metric] for model in model_names]
        
        bars = ax.bar(range(len(model_names)), metric_values, 
                      color=['#2ca02c', '#ff7f0e', '#1f77b4', '#d62728'])
        ax.set_title(f'{metric.capitalize()} Comparison', fontsize=14, fontweight='bold')
        ax.set_xlabel('Model Configuration', fontsize=12)
        ax.set_ylabel(metric.capitalize(), fontsize=12)
        ax.set_xticks(range(len(model_names)))
        ax.set_xticklabels(['Full', 'Remove B2', 'Remove B3', 'Remove B4'], 
                          rotation=15, ha='right')
        ax.set_ylim([0, 1.1])
        
        # Add value labels on bars
        for bar, value in zip(bars, metric_values):
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., height + 0.02,
                   f'{value:.3f}', ha='center', va='bottom', fontsize=10, fontweight='bold')
        
        ax.grid(True, alpha=0.3, axis='y')
    
    # Plot parameter comparison
    ax = axes[7]
    param_values = [ablation_results[model]['total_params'] for model in model_names]
    bars = ax.bar(range(len(model_names)), param_values, 
                  color=['#2ca02c', '#ff7f0e', '#1f77b4', '#d62728'])
    ax.set_title('Model Parameters Comparison', fontsize=14, fontweight='bold')
    ax.set_xlabel('Model Configuration', fontsize=12)
    ax.set_ylabel('Number of Parameters', fontsize=12)
    ax.set_xticks(range(len(model_names)))
    ax.set_xticklabels(['Full', 'Remove B2', 'Remove B3', 'Remove B4'], 
                       rotation=15, ha='right')
    ax.set_yscale('log')  # Log scale for better visualization
    
    # Add value labels
    for bar, value in zip(bars, param_values):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height * 1.1,
               f'{value:,}', ha='center', va='bottom', fontsize=10, fontweight='bold')
    
    ax.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.savefig('ablation_study_one_block_removal.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Calculate performance drop
    print("\n" + "="*100)
    print("PERFORMANCE DROP ANALYSIS (Compared to Full Model)")
    print("="*100)
    
    full_metrics = ablation_results[model_names[0]]['overall_metrics']
    
    for model_name in model_names[1:]:
        print(f"\n{model_name}:")
        current_metrics = ablation_results[model_name]['overall_metrics']
        for metric in metrics_names:
            drop = (full_metrics[metric] - current_metrics[metric]) * 100
            print(f"  {metric.capitalize()}: {current_metrics[metric]:.4f} (Drop: {drop:.2f}%)")
    
    # Identify most important block
    print("\n" + "="*100)
    print("BLOCK IMPORTANCE RANKING")
    print("="*100)
    
    block_importance = {}
    for model_name in model_names[1:]:
        if 'Block 2' in model_name:
            block = 'Block 2'
        elif 'Block 3' in model_name:
            block = 'Block 3'
        elif 'Block 4' in model_name:
            block = 'Block 4'
        else:
            continue
        
        current_metrics = ablation_results[model_name]['overall_metrics']
        acc_drop = (full_metrics['accuracy'] - current_metrics['accuracy']) * 100
        block_importance[block] = acc_drop
    
    # Sort by importance
    sorted_blocks = sorted(block_importance.items(), key=lambda x: x[1], reverse=True)
    print("\nMost important to least important blocks (based on accuracy drop when removed):")
    for i, (block, drop) in enumerate(sorted_blocks, 1):
        print(f"  {i}. {block}: {drop:.2f}% accuracy drop")
    
    return results_df

# Cell 7: Main Function for Ablation Study
def main():
    """Main function to run ablation study"""
    print("OCT Classification - Ablation Study (Remove ONE Block at a Time)")
    print("=" * 70)
    
    # UPDATE THIS PATH to your frame-level data directory
    FRAME_LEVEL_DATA_PATH = "/home/tanvirdell3/Downloads/caserel-master/Publically available Dataset/Segmentation Sharpen filter"
    
    if not os.path.exists(FRAME_LEVEL_DATA_PATH):
        print(f"Data path {FRAME_LEVEL_DATA_PATH} does not exist!")
        return
    
    # Run ablation study
    ablation_results = run_ablation_study(
        FRAME_LEVEL_DATA_PATH,
        img_size=64,
        epochs=50,
        batch_size=8
    )
    
    # Visualize results
    if ablation_results:
        results_df = visualize_ablation_results(ablation_results)
        
        # Save results to CSV
        results_df.to_csv('ablation_study_one_block_removal.csv', index=False)
        print("\nResults saved to 'ablation_study_one_block_removal.csv'")
        
        # Save detailed results
        with open('ablation_study_one_block_removal_detailed.txt', 'w') as f:
            f.write("ABLATION STUDY DETAILED RESULTS (Remove ONE Block at a Time)\n")
            f.write("="*80 + "\n\n")
            for model_name, results in ablation_results.items():
                f.write(f"{model_name}\n")
                f.write(f"Parameters: {results['total_params']:,}\n")
                f.write("-"*40 + "\n")
                for metric, value in results['overall_metrics'].items():
                    if metric != 'confusion_matrix':
                        f.write(f"{metric}: {value:.4f}\n")
                f.write(f"\nConfusion Matrix:\n{results['overall_metrics']['confusion_matrix']}\n")
                f.write("\n" + "="*80 + "\n\n")
        
        print("Detailed results saved to 'ablation_study_one_block_removal_detailed.txt'")

# Run the main function
if __name__ == "__main__":
    main()
```
