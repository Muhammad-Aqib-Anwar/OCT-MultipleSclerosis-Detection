```python
import pandas as pd
import glob

# Define your desired column order
desired_columns = ['ilm', 'nflgcl', 'iplinl', 'inlopl', 'oplonl', 'isos', 'rpe']

# Find all xlsx files in directory
file_paths = glob.glob('*.xlsx')

# List to store all dataframes
all_dfs = []

# Read and process each file
for file_path in file_paths:
    df = pd.read_excel(file_path)
    
    # Check if all required columns exist in the file
    missing_cols = [col for col in desired_columns if col not in df.columns]
    
    if missing_cols:
        print(f"Warning: File {file_path} is missing columns: {missing_cols}")
        # Add missing columns with NaN values
        for col in missing_cols:
            df[col] = None
    
    # Reorder columns and select only desired columns
    df_reordered = df[desired_columns]
    all_dfs.append(df_reordered)

# Merge all dataframes
merged_df = pd.concat(all_dfs, ignore_index=True)

# Save to output file
merged_df.to_excel('merged_output.xlsx', index=False)

print("Files merged successfully!")
```

```python


# Assuming your dataframe is named 'df1_ms'
df = df2_control.copy()

# List of layer columns (excluding Sample MS and Frames)
layer_columns = ['ilm', 'nflgcl', 'iplinl', 'inlopl', 'oplonl', 'isos', 'rpe']

# Method 1A: Using groupby and head to get first 2 rows per sample
first_two_rows = df.groupby('Sample HC').tail(3)

# Calculate mean for each layer and rename columns
layer_means = first_two_rows[layer_columns].mean().to_frame().T
layer_means.columns = [f"{col}_mean" for col in layer_means.columns]

print("Mean of each layer (first 2 rows per sample):")
print(layer_means)

# Calculate overall mean across all layers
overall_mean = first_two_rows[layer_columns].mean().mean()
print(f"\nOverall mean across all layers: {overall_mean}")

# Calculate mean per sample and rename columns
mean_per_sample = first_two_rows.groupby('Sample HC')[layer_columns].mean()
mean_per_sample.columns = [f"{col}_mean" for col in mean_per_sample.columns]

print("\nMean per sample (first 2 rows):")
print(mean_per_sample)

# Add overall mean as a new row
results_df_HC = mean_per_sample.copy()
results_df_HC.loc['Overall_Mean'] = layer_means.iloc[0]

print("\nMean per sample with overall mean:")
print(results_df_HC)
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Assuming your dataframes: sample_stats_control (14 samples) and sample_stats_MS (21 samples)

print("="*80)
print("ONE-WAY ANOVA: MS vs CONTROL FOR EACH RETINAL LAYER")
print("="*80)

layers = ['ilm', 'nflgcl', 'iplinl', 'inlopl', 'oplonl', 'isos', 'rpe']

one_way_results = []

for layer in layers:
    # Extract mean values
    # control_means = sample_stats_control[f'{layer}_mean'].dropna().values
    # ms_means = sample_stats_MS[f'{layer}_mean'].dropna().values
    ms_means = results_df_MS[f'{layer}_mean'].dropna().values
    control_means = results_df_HC[f'{layer}_mean'].dropna().values
    # Combine data
    all_data = np.concatenate([control_means, ms_means])
    group_labels = np.concatenate([['Control'] * len(control_means), 
                                   ['MS'] * len(ms_means)])
    
    # Create DataFrame for statsmodels
    data_df = pd.DataFrame({
        'thickness': all_data,
        'group': group_labels
    })
    
    # 1. Perform One-way ANOVA
    from statsmodels.formula.api import ols
    from statsmodels.stats.anova import anova_lm
    
    model = ols('thickness ~ group', data=data_df).fit()
    anova_table = anova_lm(model, typ=2)
    
    # Extract p-value
    p_value_anova = anova_table['PR(>F)'][0]
    
    # 2. Also do F-test (equivalent to ANOVA for 2 groups)
    f_stat, p_value_ftest = stats.f_oneway(control_means, ms_means)
    
    # Calculate group statistics
    control_mean = np.mean(control_means)
    control_sd = np.std(control_means, ddof=1)
    ms_mean = np.mean(ms_means)
    ms_sd = np.std(ms_means, ddof=1)
    
    # Calculate effect size (Eta squared)
    ss_between = anova_table['sum_sq'][0]
    ss_total = ss_between + anova_table['sum_sq'][1]
    eta_squared = ss_between / ss_total
    
    # Partial eta squared
    partial_eta_squared = ss_between / (ss_between + anova_table['sum_sq'][1])
    
    one_way_results.append({
        'Layer': layer.upper(),
        'Control_Mean±SD': f"{control_mean:.1f} ± {control_sd:.1f}",
        'MS_Mean±SD': f"{ms_mean:.1f} ± {ms_sd:.1f}",
        'F_statistic': f"{f_stat:.3f}",
        'p_value_ANOVA': p_value_anova,
        'p_value_Ftest': p_value_ftest,
        'Eta_squared': eta_squared,
        'Partial_Eta_squared': partial_eta_squared,
        'Interpretation': 'Large' if eta_squared >= 0.14 else 'Medium' if eta_squared >= 0.06 else 'Small'
    })

# Create results DataFrame
one_way_df = pd.DataFrame(one_way_results)

# Add significance markers
def add_sig(p):
    if p < 0.001:
        return f"{p:.4f} ***"
    elif p < 0.01:
        return f"{p:.4f} **"
    elif p <= 0.0559:
        return f"{p:.4f} *"
    else:
        return f"{p:.4f}"

one_way_df['p_value_sig'] = one_way_df['p_value_ANOVA'].apply(add_sig)

print("\nOne-way ANOVA Results:")
print(one_way_df[['Layer', 'Control_Mean±SD', 'MS_Mean±SD', 
                  'F_statistic', 'p_value_sig', 'Eta_squared', 'Interpretation']].to_string(index=False))

print("\n" + "="*80)
print("Effect Size Interpretation (Eta squared η²):")
print("-"*80)
print("η² < 0.01: Negligible effect")
print("0.01 ≤ η² < 0.06: Small effect")
print("0.06 ≤ η² < 0.14: Medium effect")
print("η² ≥ 0.14: Large effect")
```
